```python
print("24BAD050-JOHIYA SRI S")
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r"/kaggle/input/ecommerce-data/data.csv",encoding="latin1")
print(df.head())
print(df.tail())
print(df.describe())
print(df.info())
print(df.isnull().sum())
df['Sales'] = df['Quantity'] * df['UnitPrice']
top_products = df.groupby('Description')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure()
top_products.plot(kind='bar')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.title('Top 10 Products by Sales')
plt.xticks(rotation=90)
plt.show()
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
daily_sales = df.groupby(df['InvoiceDate'].dt.date)['Sales'].sum()
plt.figure()
plt.plot(daily_sales)
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.title('Daily Sales Trend')
plt.xticks(rotation=45)
plt.show()

```

```python
print("24BAD050-JOHIYA SRI S")
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r"/kaggle/input/housing-price-prediction/Housing.csv",encoding="latin1")
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
print(df.isnull().sum())
plt.figure()
plt.scatter(df['area'], df['price'])
plt.xlabel('Area (sq ft)')
plt.ylabel('Price')
plt.title('Area vs Price')
plt.show()
plt.figure()
plt.scatter(df['bedrooms'], df['price'])
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.title('Bedrooms vs Price')
plt.show()
plt.figure()
plt.scatter(df['bathrooms'], df['price'])
plt.xlabel('Number of Bathrooms')
plt.ylabel('Price')
plt.title('Bathrooms vs Price')
plt.show()
num_df = df[['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']]
corr = num_df.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Housing Features')
plt.show()
corr['price'].sort_values(ascending=False)






```

```python
print("24BAD050-JOHIYA SRI S")
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r"/kaggle/input/customer-personality-analysis/marketing_campaign.csv",sep="\t")
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
print(df.isnull().sum())
df['Age'] = 2024 - df['Year_Birth']
plt.figure()
plt.hist(df['Age'], bins=15)
plt.xlabel('Age')
plt.ylabel('Number of Customers')
plt.title('Age Distribution of Customers')
plt.show()
plt.figure()
plt.boxplot(df['Income'].dropna())
plt.ylabel('Income')
plt.title('Income Distribution of Customers')
plt.show()
avg_wine_spending = df.groupby('Education')['MntWines'].mean()
plt.figure()
avg_wine_spending.plot(kind='bar')
plt.xlabel('Education Level')
plt.ylabel('Average Wine Spending')
plt.title('Average Wine Spending by Education')
plt.xticks(rotation=45)
plt.show()
df['Total_Spending'] = df[['MntWines','MntFruits','MntMeatProducts',
                            'MntFishProducts','MntSweetProducts',
                            'MntGoldProds']].sum(axis=1)
plt.figure()
plt.boxplot(df['Total_Spending'])
plt.ylabel('Total Spending')
plt.title('Overall Spending Pattern of Customers')
plt.show()
df.groupby('Marital_Status')[['Income', 'Total_Spending']].mean()


```
